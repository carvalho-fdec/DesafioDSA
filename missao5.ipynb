## Análise Geral de Compras

```python
# Implemente aqui sua solução

# Número de itens exclusivos

df2 = df['Nome do Item']
itens = set()
itens2 = set()

for item in df2:
    aux = item.split(',')
    for i in aux:
        itens.add(i.strip())

   
print(f'Total de {len(itens)} itens')

for item in df2:
    aux = item.split(',')
    if len(aux) < 2:
        itens2.add(aux[0])

print(f'Total de {len(itens2)} itens')

itens2

```

```python
# contagem pelo item ID
df3 = df['Item ID']
print(f'Tamanho no item Id: {df3.count()}')

df4 = set(df3)
print(f'Tamanho no item Id apos o set: {len(df4)}')

print(f'Tamanho no item Id com unique: {len(df3.unique())}')
```

```python
# Preço médio de compra
media = df['Valor'].mean()
print(media)
print(f'Valor médio: {format(media,".2f")}')
```

```python
# Número total de compras
total = df.count()
print(f'Total de compras: {total[0]}')
```

## Consumidores Mais Populares (Top 5)

```python
# Implemente aqui sua solução
# Login
# Número de compras
# Preço médio de compra
# Valor Total de Compra
# Itens mais populares

df_maispop = df.groupby(["Login"]).count()["Valor"].rename("Num Compras")
df_maispop_media = df.groupby(["Login"]).mean()["Valor"].rename("Preço medio")
df_maispop_total = df.groupby(["Login"]).sum()["Valor"].rename("Valor total")
#df_maispop_itens = df.groupby(["Login"]).["Nome do Item"].rename("Itens")
#df_maispop = pd.DataFrame(df_maispop)
df_maispop = pd.merge(df_maispop, df_maispop_media, on="Login")
df_maispop = pd.merge(df_maispop, df_maispop_total, on="Login")
#df_maispop = pd.merge(df_maispop, df_maispop_itens, on="Login")
df_maispop.sort_values("Num Compras", ascending=False).head()
```

## Itens Mais Populares

```python
# Implemente aqui sua solução 
#ID do item
#Nome do item
#Número de compras
#Preço do item
#Valor Total de Compra
#Itens mais lucrativos

df_itens_qtde = df.groupby(["Item ID","Nome do Item"]).count()["Valor"].rename("Num de Compras")
df_itens_media = df.groupby(["Item ID", "Nome do Item"]).mean()["Valor"].rename("Preço medio")
df_itens_valortot = df.groupby(["Item ID", "Nome do Item"]).sum()["Valor"].rename("Total")

df_itens = pd.merge(df_itens_qtde, df_itens_media, on=("Item ID","Nome do Item"))
df_itens = pd.merge(df_itens, df_itens_valortot, on=("Item ID","Nome do Item"))
df_itens.sort_values("Num de Compras", ascending=False).head(30)

```
